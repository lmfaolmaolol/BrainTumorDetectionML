```python
# TRAINING PHASE


from google.colab import drive

drive.mount('/content/gdrive') # connecting google colab to google drive

ROOT_DIR = '/content/gdrive/My Drive/brain tumor/brain tumor detection.v2-mahitha.yolov8' # path to the dataset

!pip install ultralytics

import os

import locale

from ultralytics import YOLO

learning_rates=[0.01,0.001,0.0001]
batch_sizes=[16,32]
epochs=[50,100]

# to check for the best combination of hyper parameters, I ran 3 for loops to check and train in all combinations of the hyperparameters

for lr in learning_rates:
  for bs in batch_sizes:
    for ep in epochs:
      model = YOLO("yolov8n.yaml")
      results = model.train(data=os.path.join(ROOT_DIR, "data.yaml"), epochs=ep, batch=bs, lr0=lr)
      # to save the training data in google drive
      locale.getpreferredencoding = lambda: "UTF-8"

      !scp -r /content/runs '/content/gdrive/My Drive/brain tumor/brain tumor detection.v2-mahitha.yolov8'
```

```python
#CODE TO GET FINAL TRAINED MODEL

from google.colab import drive

drive.mount('/content/gdrive')

ROOT_DIR = '/content/gdrive/My Drive/brain tumor/brain tumor detection.v2-mahitha.yolov8'

!pip install ultralytics

import os

import locale

from ultralytics import YOLO

# After checking the reults of the previous cases, learning rate 0.001 and batch size 16 performed the best,
# so i trained another model with same learning rate and batch size but an epoch of 200

learning_rates=[0.001]
batch_sizes=[16]
epochs=[200]

for lr in learning_rates:
  for bs in batch_sizes:
    for ep in epochs:
      model = YOLO("yolov8n.yaml")
      results = model.train(data=os.path.join(ROOT_DIR, "data.yaml"), epochs=ep, batch=bs, lr0=lr)

      locale.getpreferredencoding = lambda: "UTF-8"

      !scp -r /content/runs '/content/gdrive/My Drive/brain tumor/brain tumor detection.v2-mahitha.yolov8'




```
